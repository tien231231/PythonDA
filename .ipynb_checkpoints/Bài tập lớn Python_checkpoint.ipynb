<h1>Báo cáo số liệu về Covid - 19<h1>

```python
# Cài thư viện
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.available
import seaborn as sns
%matplotlib inline
```

```python
# Đọc dữ liệu
confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveries_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
print(confirmed_df.shape) 
print(deaths_df.shape)
print(recoveries_df.shape)
```

```python
confirmed_df.head()
```

```python
deaths_df.head()
```

```python
recoveries_df.head()
```

<h3>Làm sạch dữ liệu<h3>

<h4>Tối giản tên cho các cột<h4>

```python
col=confirmed_df.columns[-1]

confirmed_df.rename(columns = {'Province/State' : 'Province', 'Country/Region' : 'Country', col : 'Current'},inplace = True)
deaths_df.rename(columns = {'Province/State' : 'Province', 'Country/Region' : 'Country', col : 'Current'},inplace = True)
recoveries_df.rename(columns = {'Province/State' : 'Province', 'Country/Region' : 'Country', col : 'Current'},inplace = True)
confirmed_df.head(1)
```

```python
deaths_df.head(1)
```

```python
recoveries_df.head(1)
```

<h4>Tạo dataframe mới gộp các nước riêng biệt

```python
confirms = pd.DataFrame(confirmed_df.groupby('Country').sum())
confirms.reset_index(inplace = True)
confirms.head(2)
```

```python
deaths= pd.DataFrame(deaths_df.groupby('Country').sum())
deaths.reset_index(inplace = True)
deaths.head(2)
```

```python
recoveries= pd.DataFrame(recoveries_df.groupby('Country').sum())
recoveries.reset_index(inplace = True)
recoveries.head(2)
```

```python
col = confirms['Country']
confirms.drop(['Lat','Long'],axis=1,inplace=True)
confirms.head(2)
```

```python
deaths.drop(['Lat','Long'],axis=1,inplace=True)
deaths.head(2)
```

```python
recoveries.drop(['Lat','Long'],axis=1,inplace=True)
recoveries.head(2)
```

```python
# Xuất dữ liệu ra csv
confirms.to_csv('confirm_data.csv')
deaths.to_csv('death_data.csv')
recoveries.to_csv('recovery_data.csv')
```

<h4>Tổng hợp<h4>

```python
print("Ca nhiễm:" , confirms.iloc[:,-1].sum())
print("Tử vong:" , deaths.iloc[:,-1].sum())
print("Hồi phục:" , recoveries.iloc[:,561].sum()) # Dữ liệu hồi phục đến 04/08/2021 bị ngừng cập nhật
```

```python
print("Tỷ lệ tử vong: ",100*deaths.iloc[:,-1].sum()/confirms.iloc[:,-1].sum(), "%")
print("Tỷ lệ hồi phục: ",100*recoveries.iloc[:,561].sum()/confirms.iloc[:,561].sum(), "%") # Dữ liệu hồi phục tính đến ngày 04/08/2021
```

<h4>Top 10 nước<h4>

```python
confirms_data = confirms[['Country','Current']].sort_values('Current',ascending = False)
deaths_data = deaths[['Country','Current']].sort_values('Current',ascending = False)
recoveries_data = recoveries[['Country','8/4/21']].sort_values('8/4/21',ascending = False)
```

```python
# Confirmed Cases
sns.set(font_scale=1.5)
plt.figure(figsize=(10,5))
fig= sns.barplot(x='Current', y='Country', data=confirms_data[:10], orient='h',color='Red')
plt.title('Total Confirmed Cases Worldwide')
fig.set(xlabel ='Number of Cases', ylabel ='Country')
plt.show()

# Death Cases
plt.figure(figsize=(10,5))
fig= sns.barplot(x='Current', y='Country', data=deaths_data[:10], orient='h',color='Black')
plt.title('Total Deaths Worldwide')
fig.set(xlabel ='Number of Cases', ylabel ='Country')
plt.show()

# Recovery Cases
plt.figure(figsize=(10,5))
fig= sns.barplot(x='8/4/21', y='Country', data=recoveries_data[:10], orient='h',color='Blue')
plt.title('Total Recovery Worldwide')
fig.set(xlabel ='Number of Cases', ylabel ='Country')
plt.show()
```

<h4>Các ca bệnh hằng ngày của các nước ĐNA<h4>

```python
vietnam_confirm = confirms[confirms.Country == 'Vietnam'].iloc[:,1:].sum().values.tolist()
laos_confirm = confirms[confirms.Country == 'Laos'].iloc[:,1:].sum().values.tolist()
cambodia_confirm = confirms[confirms.Country == 'Cambodia'].iloc[:,1:].sum().values.tolist()
thailand_confirm = confirms[confirms.Country == 'Thailand'].iloc[:,1:].sum().values.tolist()
myanmar_confirm = confirms[confirms.Country == 'Burma'].iloc[:,1:].sum().values.tolist()
philippines_confirm = confirms[confirms.Country == 'Philippines'].iloc[:,1:].sum().values.tolist()
malaysia_confirm = confirms[confirms.Country == 'Malaysia'].iloc[:,1:].sum().values.tolist()
indonesia_confirm = confirms[confirms.Country == 'Indonesia'].iloc[:,1:].sum().values.tolist()
singapore_confirm = confirms[confirms.Country == 'Singapore'].iloc[:,1:].sum().values.tolist()
timo_confirm = confirms[confirms.Country == 'Timor-Leste'].iloc[:,1:].sum().values.tolist()
brunei_confirm = confirms[confirms.Country == 'Brunei'].iloc[:,1:].sum().values.tolist()
```

```python
# Confirmed Cases
plt.figure(figsize=(16,9))
plt.plot(vietnam_confirm)
plt.plot(laos_confirm)
plt.plot(cambodia_confirm)
plt.plot(thailand_confirm)
plt.plot(myanmar_confirm)
plt.plot(philippines_confirm)
plt.plot(malaysia_confirm)
plt.plot(indonesia_confirm)
plt.plot(singapore_confirm)
plt.plot(timo_confirm)
plt.plot(brunei_confirm)

plt.title('Các ca nhiễm ở Đông Nam Á', size=25)
plt.xlabel('Số ngày từ 22/1/2020', size=20)
plt.ylabel('Số ca nhiễm', size=20)
plt.legend(['Vietnam', 'Laos','Campuchia','Thái Lan','Myanmar','Phillipines', 'Malaysia', 'Indonesia', 'Singapore', 'Timo-Leste', 'Brunei'])
plt.show()
```

Nhận xét: 
Tính từ ngày 22/01/2020 
- Số ca nhiễm bắt đầu tăng từ ngày thứ 200
- Từ ngày thứ 200 đến ngày thứ 400 số lượng người nhiễm bắt đầu tăng mạnh. Đặc biệt là Indonesia, Phillipines, Malaysia, Thái Lan.
- Từ ngày thứ 400 đến 600, các nước tăng mạnh như Indonesia, Phillipines, Malaysia, Thái Lan vẫn tăng đều. Đến ngày thứ 750 thì số người nhiễm của Việt Nam tăng đột biến.
- Từ này thứ 800 và 1000, Số ca nhiễm của các nước đứng đầu về ca nhiễm tăng chậm nhưng số ca nhiễm ở Việt Nam vẫn tăng mạnh.
- Các nước có số ca nhiễm ít nhất thường là các nước có dân số ít.
